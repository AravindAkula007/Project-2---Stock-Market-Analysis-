ARIMA (AutoRegressive Integrated Moving Average) for Time Series Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
df_arima = pd.DataFrame({
    'Date': pd.date_range(start='2020-01-01', periods=100),
    'Close': np.random.rand(100) * 100  # Replace with actual stock prices
})

df_arima.set_index('Date', inplace=True)
```

```python
#Split the data into training and testing sets
train_data_len = int(len(df_arima) * 0.8)
train = df_arima['Close'][:train_data_len]
test = df_arima['Close'][train_data_len:]
```

```python
#Fit the ARIMA model
model = ARIMA(train, order=(1, 0, 1))
model_fit = model.fit()
```

```python
predictions = model_fit.forecast(steps=len(test))
```

```python
#Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, predictions))
print('RMSE:', rmse)
```

```python
# Step 6: Plot the actual vs. forecasted prices
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data', color='blue')
plt.plot(test.index, test, label='Actual Prices', color='green')
plt.plot(test.index, predictions, label='Forecasted Prices', color='red')
plt.title('Stock Price Forecast using ARIMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

LSTM based RNN using Tensorflow

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = df[['Date', 'Close']].copy()
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Normalize the 'Close' prices
scaler = MinMaxScaler(feature_range=(0, 1))
data['Close'] = scaler.fit_transform(data['Close'].values.reshape(-1, 1))
```

```python
# Create sequences for the RNN model
def create_sequences(data, sequence_length):
    sequences = []
    labels = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        label = data[i+sequence_length]
        sequences.append(seq)
        labels.append(label)
    return np.array(sequences), np.array(labels)

# Define sequence length
sequence_length = 60  # Using the past 60 days to predict the next day

# Create sequences
X, y = create_sequences(data['Close'].values, sequence_length)
```

```python
# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape the data to fit the model (add an extra dimension for features)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
!pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout
```

```python
# Build the RNN model
model = Sequential()
```

```python
# Add an LSTM layer with Dropout regularization
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
```

```python
# Add another LSTM layer with Dropout
model.add(LSTM(units=60, activation= 'relu', return_sequences=True))
model.add(Dropout(0.3))
```

```python
# Add another LSTM layer with Dropout
model.add(LSTM(units=120, activation= 'relu'))
model.add(Dropout(0.4))
```

```python
# Add the output layer
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Make predictions on the test set
predicted_prices = model.predict(X_test)

# Inverse transform the predicted and actual prices to original scale
predicted_prices = scaler.inverse_transform(predicted_prices)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Plot the results
plt.figure(figsize=(14,5))
plt.plot(data.index[-len(actual_prices):], actual_prices, color='blue', label='Actual AMAZON Stock Price')
plt.plot(data.index[-len(predicted_prices):], predicted_prices, color='red', label='Predicted AMAZON Stock Price')
plt.title('AMAZON Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Define the specific future date (format: 'YYYY-MM-DD')
specific_future_date = '2024-08-31'
specific_future_date = pd.to_datetime(specific_future_date)

# Calculate the number of days to forecast
last_date = data.index[-1]
forecast_days = (specific_future_date - last_date).days

# Ensure forecast_days is a positive number
if forecast_days <= 0:
    raise ValueError("The specific future date must be after the last date in the dataset.")
```

```python
# Start with the last sequence from the test set
last_sequence = X_test[-1]

# Initialize a list to store the future predictions
future_predictions = []

# Predict future prices up to the specific future date
for _ in range(forecast_days):
    # Reshape the last sequence to match the model input
    last_sequence_reshaped = last_sequence.reshape(1, last_sequence.shape[0], 1)
    
    # Predict the next price
    next_price = model.predict(last_sequence_reshaped)[0]
    
    # Append the prediction to the list
    future_predictions.append(next_price)
    
    # Update the sequence by removing the first element and adding the new prediction
    last_sequence = np.append(last_sequence[1:], next_price)
```

```python
# Inverse transform the predictions to get the actual price values
future_predictions = scaler.inverse_transform(future_predictions)

# Generate dates for the future predictions
future_dates = pd.date_range(last_date, periods=forecast_days+1, inclusive='right')

# Create a DataFrame to hold the future predictions
forecast_df = pd.DataFrame(data=future_predictions, index=future_dates, columns=['Predicted Close'])
```

```python
import matplotlib.pyplot as plt

# Plot the actual and predicted prices
plt.figure(figsize=(14,5))
plt.plot(data.index, scaler.inverse_transform(data['Close'].values.reshape(-1, 1)), color='blue', label='Actual AMAZON Stock Price')
plt.plot(forecast_df.index, forecast_df['Predicted Close'], color='green', linestyle='dashed', label='Forecasted AMAZON Stock Price')
plt.title('AMAZON Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
